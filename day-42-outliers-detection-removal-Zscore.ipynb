## Capping

```python

upper_limit = df['Carea'].mean() + 3*df['Carea'].std()
lower_limit = df['Carea'].mean() - 3*df['Carea'].std()
```

```python
lower_limit
```

```python
df['Carea'] = np.where(
    df['Carea'] > upper_limit,
    upper_limit,
    np.where(df['Carea'] < lower_limit,
            lower_limit,
             df['Carea']
            )
) 
```
