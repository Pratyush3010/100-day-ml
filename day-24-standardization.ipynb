```python
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mergedCarea.csv')
```

```python
df = df.iloc[:, 2:]
```

df

```python
df = df.dropna()
```

```python
df.sample(5)
```

```python
df = df.drop(df[df['Carea'] < 0].index)
df
```

## Train test spilt

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop('Carea', axis=1),
                                                    df['Carea'],
                                                    test_size=0.3,
                                                    random_state=0)
```

```python
X_train.shape, X_test.shape,y_train.shape,y_test.shape
```

## Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# fit the scaler to train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.mean_
```

```python
X_train
```

```python
X_train_scaled
```

```python
# This code is used to convert scaled NumPy arrays back into Pandas DataFrames while keeping the original column names from X_train and X_test
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
# X_train_scaled = pd.DataFrame(X_train_scaled,columns= X_train.column)
# X_test_scaled = pd.DataFrame(X_test_scaled,columns= X_test.columns)
```

## Why scaling is important?

Logistic Regression is used for the Classification Problem,but here we have the regression problem as our taget is continuous not discrete so we use linear regression not logistic.

```python
from sklearn.linear_model import LinearRegression
from sklearn.cross_decomposition import PLSRegression
from sklearn.metrics import r2_score
```

```python
lr = PLSRegression()
lr_scaled = PLSRegression()
```

```python

lr.fit(X_train,y_train)
lr_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python

print("Actual",mean_squared_error(y_test,y_pred))
print("Scaled",mean_squared_error(y_test,y_pred_scaled))
```

```python

print("Actual",r2_score(y_test,y_pred))
print("Scaled",r2_score(y_test,y_pred_scaled))
```
