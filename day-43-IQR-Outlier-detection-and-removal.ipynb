```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mergedCarea.csv')
```

```python
df.head()
```

```python

df = df.drop(df[df['Carea']<0].index)
df.shape
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.distplot(df['Carea'])

plt.subplot(1,2,2)
sns.distplot(df['350'])

plt.show()
```

```python
df['Carea'].describe()
```

```python
sns.boxplot (df['Carea'])
```

```python
percentile25 = df['Carea'].quantile(0.25)
percentile75 = df['Carea'].quantile(0.75)
```

```python
percentile75
```

```python
iqr = percentile75 - percentile25
```

```python
iqr
```

```python
upper_limit = percentile75 + 1.5* iqr
```

```python
lower_limit = percentile25 - 1.5 * iqr
```

```python
print("Upper limit",upper_limit)
print('Lower limit',lower_limit)
```

## Finding Outliers

```python
df[df['Carea'] > upper_limit]
```

```python
df[df['Carea'] < lower_limit]
```

## Trimming

```python
new_df = df[df['Carea']<upper_limit]
```

```python
new_df.shape
```

```python

# Comparing

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['Carea'])

plt.subplot(2,2,2)
sns.boxplot(df['Carea'])

plt.subplot(2,2,3)
sns.distplot(new_df['Carea'])

plt.subplot(2,2,4)
sns.boxplot(new_df['Carea'])

plt.show()
```

## Capping

```python

new_df_cap = df.copy()

new_df_cap['Carea'] = np.where(
    new_df_cap['Carea'] > upper_limit,
    upper_limit,
    np.where(
        new_df_cap['Carea'] < lower_limit,
        lower_limit,
        new_df_cap['Carea']
    )
)
```

```python
new_df_cap.shape
```

```python

# Comparing

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['Carea'])

plt.subplot(2,2,2)
sns.boxplot(df['Carea'])

plt.subplot(2,2,3)
sns.distplot(new_df_cap['Carea'])

plt.subplot(2,2,4)
sns.boxplot(new_df_cap['Carea'])

plt.show()
```
